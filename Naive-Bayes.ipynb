### Naive Bayes Implementation

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
import sys
import numpy as np
```

```python
X, y = load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
df = sns.load_dataset("tips")
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Dependent and Independent Features
X = df[["total_bill", "tip", "smoker", "day", "time", "size"]]
y = df.iloc[:, 2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
X_train.head()
```

```python
# Feature Encoding (Label and OneHot Encoding)
onehot = OneHotEncoder()
label_smoker = LabelEncoder()
label_time = LabelEncoder()
```

```python
X_train["smoker"] = label_smoker.fit_transform(X_train["smoker"])
X_train["time"] = label_time.fit_transform(X_train["time"])
```

```python
X_train.head()
```

```python
X_test["smoker"] = label_smoker.transform(X_test["smoker"])
X_test["time"] = label_time.transform(X_test["time"])
```

```python
X_test.head()
```

```python
ct = ColumnTransformer(transformers=[("onehot", OneHotEncoder(drop="first"), [3])], remainder="passthrough")
```

```python
np.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

### Hyperparameter Tuning

```python
params = {
    "var_smoothing": [1e-9, 1e-8, 1e-7, 1e-6]
}
```

```python
# Grid SearchCV
grid = GridSearchCV(estimator=gnb, param_grid=params, cv=5)
```
